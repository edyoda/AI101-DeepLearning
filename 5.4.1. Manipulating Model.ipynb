```python
import keras
from keras import backend as K
import tensorflow as tf;
sess = tf.Session()
K.set_session(sess)

from keras.applications import VGG16
#Load the VGG model
vgg_conv = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
```

```python
from keras import Model;

model = Model(vgg_conv.input,vgg_conv.layers[-4].output)
model.summary()
```

```python
inp = tf.placeholder(shape=(None,224,224,3),dtype=tf.float32)
lbl = tf.placeholder(shape=(None,2),dtype=tf.float32)
```

```python
from keras import layers
finetune_out = model(inp)
y1 = layers.Flatten()(finetune_out)
y2 = layers.Dense(1024, activation='relu')(y1)
y3 = layers.Dropout(0.5)(y2)
ypred = layers.Dense(2, activation='softmax')(y3)
```
